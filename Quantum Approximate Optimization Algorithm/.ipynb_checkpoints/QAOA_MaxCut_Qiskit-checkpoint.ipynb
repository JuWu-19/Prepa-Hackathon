# Solving combinatorial optimization problems using QAOA

This notebook presents a practical implementation of noisy quantum channels and an analysis of its influence over the performance of QAOA algorithm.
The first part of this notebook is largely inspired from the Qiskit Textbook, written by IBM. 
In a second time, I implement a generalized procedure to process classical optimization of the angles to yield the best expectation value (not done in the textbook, expectation value expression was calculated and maximized analytically in the original textbook). Finally I choose an arbitrary generated graph and perform noiseless, and noisy analysis over the problem of finding the MaxCut.


##  Qiskit Implementation<a id='implementation'></a>

As the example implementation we consider the $MAXCUT$ problem on the butterfly graph of the openly avaliable IBMQ 5-qubit chip. The graph will be defined below and corresponds to the native connectivity of the device. This allows us to implement the original version of the $QAOA$ algorithm, where the cost - function $C$ and the Hamiltonian $H$ that is used to generate the state coincide. Moreover, for such a simple graph the exact cost function can be calculated analytically. To implement the circuit, we follow the notation and gate defintions from the [Qiskit Documentation](https://qiskit.org/documentation/).

###  Quantum circuit<a id="implementation_sec53"></a>  #Qiskit textbook

 We prepare the state 



$$ |\psi_1(\gamma ,\beta)\rangle = e^{ -i\beta B } e^{ -i\gamma H } |+\rangle^n.  $$



Recall that the terms are given by $B = \sum_{k \in V} X_k$ and $H = \sum_{(k,m) \in E} \frac{1}{2}\left(1 - Z_kZ_m\right)$. To generate the circuit we follow these steps:

 - We first implement 5 - Hadamard $H$ gates to generate the uniform superposition.
 

 - This is followed by $6$ Ising type gates $U_{k,l}(\gamma)$ with angle $\gamma$ along the edges $(k,l) \in E$. This gate can be expressed in terms of the native Qiskit gates as

  

$$ U_{k,l}(\gamma) = C_{u1}(-2\gamma)_{k,l}u1(\gamma)_k u1(\gamma)_l$$


  
 
 - Lastly we apply single qubit $X$ rotations $X_k(\beta)$ for every vertex $k \in V$ with $\beta$ as angle. This gate directly parametrized as $X_k(\beta) = R_x(2\beta)_k$ in Qiskit.
 

 - In the last step we measure the qubits in the computational basis, i.e. we perform a $Z$ - measurement and record the resulting bit-string $x \in \{0,1\}^5$.

### 5.1 Problem definition 

We define the cost function in terms of the butterfly graph of the superconducting chip. The graph has $n = 5$ vertices $ V = \{0,1,2,3,4\}$ and six edges $E = \{(0,1),(0,2),(1,2),(3,2),(3,4),(4,2)\}$, which will all carry the same unit weight $w_{ij} = 1$. We load an additional network package to encode the graph and plot connectivity below.

```python
# Generating the butterfly graph with 5 nodes 
n     = 5
V     = np.arange(0,n,1)
E     =[(0,1,1.0),(0,2,1.0),(1,2,1.0),(3,2,1.0),(3,4,1.0),(4,2,1.0)] 

G     = nx.Graph()
G.add_nodes_from(V)
G.add_weighted_edges_from(E)

# Generate plot of the Graph
colors       = ['b' for node in G.nodes()]
default_axes = plt.axes(frameon=True)
pos          = nx.spring_layout(G)

nx.draw_networkx(G, node_color=colors, node_size=600, alpha=1, ax=default_axes, pos=pos)
```

## First implementation: original graph from the textbook, classical optimization subroutine done, noiseless simulation

We build here the essential components necessary to perform QAOA, that is : 
-  the Quantum Circuit : General function for MaxCut problem generating a circuit according to the definition of a graph and taking as input two sets of angles to choose the number of adiabatic evolution blocks you want to run before sampling
- the cost function, associated to the MaxCut problem, necessary to analyze how did the circuit perform, i.e how good is the cost associated to the bitstring returned by the Quantum Circuit
- a function called computation, which launches and makes the sampling of the quantum circuit with a initial set of angles. The idea behind this function is that it returns the desired expectation value that we wish to maximize in terms of the angle. This is the objective function that the classical optimizer (COBYLA, SPSA, ...) intends to maximize by tuning up the angles

Now that building blocks are defined, one has to perform the classical optimization of angles to determine the best approximate bitstring that the algorithm can yield. To perform the optimization, we choose random values $(\gamma,\beta)\in [0,2\pi]^p\times[0,\pi]^p$ and use the optimizer to maximize the ouput of the function computation, which is nothing else but the expectation value desired.
One is interested to check the performance of QAOA under the action of adding adiabatic evolution blocks, so we try this optimization for various p

## Extended function, not meant to optimize the angles but to return the outcome of the circuit for a given set of angles) Can be made independent of the graph

Now, once we have the good set of angles, we want to make a check to see how is the output bitstring close to the maximum expectation value, one expects from theory that those values should be centered around the mean.
The function below does exactly the same thing as computation, the only difference is that it will now look for the best bitstring

## Implementing local noisy operations on usual computation

In the following, we present an implementation of some noisy quantum channels to analyse the performance of QAOA under these simple local noise effects.

## Refining QAOA Circuit including local noisy channels

We take back the QAOA circuit defined earlier, and add our noisy functions. The model of local noise is implemented as follows:
Each qubit is susceptible to encounter an error of the three types defined above, but we assume for simplicity that this error occurs for all qubits at the same place in the circuit, that is between the two unitaries forming the adiabatic evolution block. In the paper Characterizing local noise in QAOA circuits written by
Jeffrey Marshall, Filip Wudarski Stuart Hadfield and Tad Hogg (https://arxiv.org/pdf/2002.11682.pdf), the implementation is nearly identical, except that the error occurs between two different blocks.

```python
def noisy_computation(angles,G,backend_1=Aer.get_backend("qasm_simulator"),shots_1=10000,prob=[0.,0.,0.]): #New function similar to computation, but for the new graph
    
    l=len(angles)
    =angles[0:l:2]
    =angles[1:l:2]
    simulate=execute(QAOA_noisy(,,G,prob),backend=backend_1,shots=shots_1)
    QAOA_results=simulate.result()
    counts=QAOA_results.get_counts()
    
    avr_C       = 0

    for sample in list(counts.keys()):

    # use sampled bit string x to compute C(x)
        x         = [int(num) for num in list(sample)]
        tmp_eng   = cost_function_C(x,G)
    
    # compute the expectation value and energy distribution
        avr_C     = avr_C    + counts[sample]*tmp_eng
        
    
    
                
    Mp_sampled   = avr_C/shots_1
    
    return -Mp_sampled
```

## Simulation without noise, using a stochastic method for classical optimization

Since the range of our parameters is not so big to explore, one can search the whole space at random to find the optimal value. However, in order for the simulation to be short, the tolerance threshold must be put to a sufficient high value to avoid looking for too precise values. The stochastic method used here is called differential evolution (link below), and tolerance is fixed to 0.1. https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.differential_evolution.html?highlight=differential%20evolution#scipy.optimize.differential_evolution

```python
#optimizer = COBYLA(maxiter=1000000, tol=0.00001) 
nb_dp=5
p=[]
exp=[]
opti_angles=[]
ratio=[]
max_value=9.  #Max value of the cost function for the graph chosen above
 #Number of iterations for evolution process
"""Choose noises types : 
1. Phase flip
2.Depolarizing noise 
3. Bit flip
4. To be added : Phase damping """
"""Choose the probability of having those errors"""
proba_phase_flip=0.0
proba_depolarizing_noise=0.0
proba_bit_flip=0.0
prob=[proba_phase_flip, proba_depolarizing_noise, proba_bit_flip]

for j in range(1,nb_dp+1):
    print("optimization for ",j,"adiabatic evolution block(s)")
    angles=[]
    min_bound=[]
    max_bound=[]

    for i in range(j):
        a=rand.uniform(0,2*)
        b=rand.uniform(0,)
        angles.append(a)
        angles.append(b)
        min_bound.append(0.)
        min_bound.append(0.)
        max_bound.append(2*)
        max_bound.append()
        
        
    p.append(j)
    boundaries=[]

    for i in range(2*j):
        boundaries.append((min_bound[i],max_bound[i]))
        
#angles2=[rand.uniform(0,2*),rand.uniform(0,),rand.uniform(0,2*),rand.uniform(0,)]
#Opti_Result = minimize(fun=computation_graph2,x0=angles2,args=(G2,Aer.get_backend("qasm_simulator"),10000),method= 'SLSQP',
                        #bounds=((0,2*),(0,),(0,2*),(0,)))
    Opti_Result=differential_evolution(noisy_computation,bounds=boundaries,args=(G2,Aer.get_backend("qasm_simulator"),10000,prob),tol=0.1)
    opti_angles.append(Opti_Result.x)
    exp.append(-Opti_Result.fun)
    ratio.append(-Opti_Result.fun/9.0)
    #exp.append(-expectation)
    #opti_angles.append(opti_angles_1)
    #ratio.append(-expectation/max_value)
print("Optimization done")
```
